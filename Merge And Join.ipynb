### `Merge()` And `Joins()` 

```python
import pandas as pd
```

```python
Restaurant = pd.DataFrame(
    {
        'restaurant_id':['R001', 'R002', 'R003', 'R004', 'R005', 'NaN', 'R007'],
        'restaurant_name':['Dominos Pizza', 'KFC', 'McDonalds', 'Pizza Hut', 'Subway', 'Unknown Restaurant', 'Taco Bell'],
        'cuisine_type':['Italian', 'American', 'Fast Food', 'Italian', 'Fast Food', 'NaN', 'Mexican'],
        'rating':[4.2, 4.0, 3.8, 4.1, 4.3, 3.5, 4.0]
    }
)
Restaurant
```

```python
Order = pd.DataFrame(
    {
        'order_id':['0001', '0002', '0003', '0004', '0005', '0006', '0007', '0008', '0009'],
        'restaurant_id':['R001', 'R002', 'NaN', 'R001', 'R005', 'NaN', 'R007', 'R999', 'R004'],
        'customer_name':['Rahul', 'Priya', 'Amit', 'Sneha', 'Rohan', 'Kavya', 'Arjun', 'Pooja', 'Vikash'],
        'order_amount':[450.0, 320.0, 280.0, 520.0, 'NaN', 380.0, 420.0, 150.0, 350.0],
        'order_date':['2024-01-15', '2024-01-16', '2024-01-17', '2024-01-18', '2024-01-19', '2024-01-20', '2024-01-21', '2024-01-22', '2024-01-23']
        
    }
)
Order
```

### `Inner Join`

```python
df = pd.merge(Restaurant, Order, on ='restaurant_id', how = "inner")
df
```

### `Left Join`

```python
df1 = pd.merge(Restaurant, Order, on ='restaurant_id', how = "left")
df1
```

### `Right Join`

```python
df2 = pd.merge(Restaurant, Order, on ='restaurant_id', how = "right")
df2
```

### `Outer Join`

```python
df3 = pd.merge(Restaurant, Order, on ='restaurant_id', how = "outer")
df3.head()
```

## Merge Data Base on index 

`Join()`

```python
df4 = pd.merge(Restaurant, Order, left_index=True,right_index=True,how='inner')
df4
```

```python
df4 = pd.merge(Restaurant, Order, left_index=True,right_index=True,how='inner', suffixes=('_R','_O'))
df4
```

```python
df5 = pd.merge(Restaurant, Order, left_index=True,right_index=True,how='left', suffixes=('_R','_O'))
df5
```

```python
df6 = pd.merge(Restaurant, Order, left_index=True,right_index=True,how='right', suffixes=('_R','_O'))
df6
```

```python
df7 = pd.merge(Restaurant, Order, left_index=True,right_index=True,how='outer', suffixes=('_R','_O'))
df7
```

### . `concat()`  Data Stacking

```python
Data1 = pd.DataFrame({
    'ID' : [1,2,3,4,5],
    'Name':['Amit', 'Bina', 'Chetan', 'Deepa', 'Esha'],
    'Age' :[21,22,23,21,24],
    'Score':[85,90,88,92,95]
})

Data1
```

```python
Data2 = pd.DataFrame({
    'ID' : [6,7,8,9,10,11,12],
    'Name':['Farahan', 'Ved', 'Ayush', 'Hemant', 'Ravi', 'Raju', 'Ankur'],
    'Age':[23,22,21,24,25,23,22],
    'Score':[87,91,85,89,93,90,88],
    'City' :['Delhi', 'Mumbai', 'Pune', 'Delhi', 'Goa', 'Chennai', 'Pune']
})

Data2
```

### Verticle Stacking

```python
df8 = pd.concat([Data1, Data2], axis =0, ignore_index = True)
df8
```

### Horizontal Stacking

```python
df9 = pd.concat([Data1, Data2], axis =1)
df9
```
